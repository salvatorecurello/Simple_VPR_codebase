```python
latitude = []
longitude = []
for filename in filenames:
  parts = filename.split("@")
  latitude.append(parts[5][1:])
  longitude.append(parts[6][3:])
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Project/Simple_VPR_codebase/utils/barcelona_coordinates.csv")
```

```python
df = df.astype({'Latitudes':'float','Longitudes':'float'})
```

```python
import plotly.express as px

fig = px.scatter_mapbox(df, lat="Latitudes", lon="Longitudes", color_discrete_sequence=["fuchsia"], zoom=13, height=1000)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
